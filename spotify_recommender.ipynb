### Extraer artistas sin repeticiones

```python
# 1.Extraer los artistas correspondientes a las top-20 tracks (sin repetición)
ids_artists = []
print('Artistas en mi top 20')
print('*-*-*-*-*-*-*-*-*-*-*')

for item in top20['items']:
    # Accedemos a la info del track 1 y extramos el artista y del artista extraemos el id
    artist_id = item['artists'][0]['id']
    artist_name = item['artists'][0]['name']
    print(artist_id, ':', artist_name)
    ids_artists.append(artist_id)
```

```python
# Depurar lista para evitar artistas repetidos
ids_artists = list(set(ids_artists))
print('Número de artistas (sin repeticiones):', {len(ids_artists)})
```

### Ampliar listado con Artistas Recomendados

```python
ids_artists
```

```python
# Por cada artista top, se buscarán los artistas similares y se añadirán
print('Artistas Similares\n*-*-*-*-*-*-*-*-*-*\n')

ids_similar_artists =[]

for artist_id in ids_artists:
    artists = sp.artist_related_artists(artist_id)['artists']
    for item in artists:
        artist_id = item['id']
        artist_name = item['name']
        print(artist_id, ':', artist_name)
        ids_similar_artists.append(artist_id)
```

```python
# Extend agrega cada uno de los elementos de la lista a ids_artists (mi top 20)
ids_artists.extend(ids_similar_artists)
# Depurar artistas repetidos
ids_artists = list(set(ids_artists))
print('Número de Artistas (sin repeticiones):', len(ids_artists))
```

### Ampliar el listado con artistas con nuevos lanzamientos

```python
# Artista de cualquier genero musical
print('Artistas con nuevos lanzamientos\n*-*--*-*-*-*-*-*-*-*-*-*-*-*-*-*\n')

# Artistas sin relación (con mi top20 y mis relacionados), son solos que esten lanzando nueva musica
new_releases = sp.new_releases(limit=20)['albums']

for item in new_releases['items']:
    artist_id = item['artists'][0]['id'] # [0] porque puede haber varios artistas
    artist_name = item['artists'][0]['name'] 
    album_name = item['name'] # Nombre del album, puramente informativo
    release_date = item['release_date'] # Fecha de lanzamiento, puramente informativo
    print(artist_id, '.:.', artist_name, "// ", album_name, "//", release_date)
    # Agregamos los id a la lista que habias creado con anterioridad
    ids_artists.append(artist_id)
    
# Depurar lista para evitar repeticiones
ids_artists = list(set(ids_artists))
print('Número de artistas (sin repeticiones): ', len(ids_artists))
```

### Obtener el listado de albums de cada uno de los anteriores artistas.
Se limitará únicamente 1 album (limit=1), para evitar tener un listado gigantesco

```python
# listado de id de albumes

id_albums = []
num_artists = len(ids_artists)

for i, id_artist in enumerate(ids_artists):
    print(f'Procesando artista {i+1} de {num_artists}...') # es desde 0 por eso desde el inicio se le suma 1 para que desde pantalla aparezca el 1
    albums = sp.artist_albums(id_artist, limit=1)
    for album in albums['items']:
        id_albums.append(album['id'])
print('¡Listo!')
```

Listado de 3 canciones de los albumes

```python
id_tracks = []
num_albums = len(id_albums)

for i, id_album in enumerate(id_albums):
    print(f'Procesando tracks {i+1} de {num_albums}...') # es desde 0 por eso desde el inicio se le suma 1 para que desde pantalla aparezca el 1
    album_tracks = sp.album_tracks(id_album, limit=3)
    # otra manera de hacer agregar a la lista
    for track in album_tracks['items']:
        id_tracks.append(track['id'])
print(f'¡Listo! Número total de tracks precandidatos realizado: {len(id_tracks)}')
```

### Extraer los audio_features para cada track
En total se añadiran n canciones y todas se añadira a un dataframe.

```python
# Lsita de nombre de canciones
track_names = []
# lista de caracteristicas
features = []
# número de tracks
ntracks = len(id_tracks)

# Desde la lista de id_tracks agregar los nombre y las caracteristica a las nuevas listas
for i, track_id in enumerate(id_tracks):
    print(f'Procesando track {i+1} de {ntracks}...')
    track_name = sp.track(track_id)['name']
    audio_features = sp.audio_features(track_id)
    
    # No incluir pista sin features establecidas por spotify
    if audio_features[0] != None:
        track_names.append(track_name)
        features.append(audio_features[0])
print('¡Listo!')
```

### Escalar los datos

```python
# Extraer sólo los features en formato numpy array
# iloc[ todas las filas y apartir de la columna 1 ]
top20_mtx = top20_df.iloc[:,1:].values
# iloc[ todas las filas y apartir de la columna 1 ]
candidatos_mtx = candidatos_df.iloc[:,1:].values
```

```python
from sklearn.preprocessing import StandardScaler

# Estandarizar cada columna de features: mu=0, sigma=1
# Ya que, cada caracteristica tiene una escala diferente

# Intanciamos clase
scaler = StandardScaler()
# Usamos el Pipeline para escalar datos
# Creamos un nuevo objeto donde aplicaremos el genereador y el transformer
t20_scaled = scaler.fit_transform(top20_mtx)
# Creamos un nuevo objeto donde aplicaremos el genereador y el transformer
can_scaled = scaler.fit_transform(candidatos_mtx)
```

```python
# Para comprobar obtenemos el valor medio y standar de los tracks
print(t20_scaled.mean(axis=0))
print(t20_scaled.std(axis=0))

# Medias son iguales a 0 y desviación standar a 1; lo que comprueba que son muy iguales
```

### Normalizar los Datos
Esto es para que los vectores tengan el mismo tamaño (tengan el tamaño de 1)

```python
import numpy as np

# Normalizar cada vector de caracteristicas (es decir por filas)

# Magnitutes de cada vector (o pista)
# Los elementos del vector se elevan al cuadrado y luego todos se suman y se optine la raiz cuadrada
t20_norm = np.sqrt((t20_scaled * t20_scaled).sum(axis=1))
can_norm = np.sqrt((can_scaled * can_scaled).sum(axis=1))

# Normalización
nt20 = t20_scaled.shape[0] # 
ncan = can_scaled.shape[0]
# Normalizacion = Vector / Magnitud se hace un rechape con el (shape anterior y 1)
t20 = t20_scaled/t20_norm.reshape(nt20, 1)
can = can_scaled/can_norm.reshape(ncan, 1)

print(np.sqrt((t20 * t20).sum(axis=1)))
print(np.sqrt((can * can).sum(axis=1)))
```

```python
from sklearn.metrics.pairwise import linear_kernel

# Calcular similitudes del coseno entre cada top20 y cada una de las pistas candidatas
cos_sim = linear_kernel(t20, can)
cos_sim.shape
```

```python
# Ejemplo: ¿Qué tanto se parece una pista candidata con una del top20?

# Grado de similitud
print(cos_sim[6,270])
print(cos_sim[3,24])
```

```python
# Obtener candidatos para una pista dada

# Dada una pista del top20, extraer candidatos.
# Usando "cos_sim" y siempre y cuando superen un umbral de similitus

# (pocision en el bucle, resultado linear kernel, num. candidatos maximos,grado de similitud(umbral))
def obtener_candidatos(pos, cos_sim, ncands, umbral=0.8):
    # *******CADA CANDIDATO PARA POR LA CONDICIÓN*********************
    # Obtener todas las pistas candidatas por encima de un umbral
    #              [pocisión y toda la lista]
    idx = np.where(cos_sim[pos,:]> umbral)[0] 
    
    # *****SE ORGANIZA LA LISTA DE CANDIDATOS POR CANCIÓN*************
    # organizarlas de forma de descendentes (por similitudes de mayor a menor)
    idx = idx[np.argsort(cos_sim[pos,idx])[::-1]] # [::-1] por defecto argsort es de menor a mayor
    
    # Si la lista de IDX es mayor que la lista NCANDS
    if len(idx) >= ncands:
        # Solo mostrar las 5 primeras
        cands = idx[0:ncands]
    else:
        # De lo contrario muestra las cands
        cands = idx
        # Retorna las cands
    return cands
```

```python
# Ejemplo de uso
for i in range(20):
    # Este CANDS es igual al CANDS que se retorno anteriormente
    cands = obtener_candidatos(i, cos_sim, 5)
    #                                                         Grado de silimitud
    #                            lista de candidatas       cos_sim[top y candidatas]
    print(f'{i} ==> pistas candidatas: {cands}, similitudes: {cos_sim[i, cands]}')
```

## Creación de playlist con listado de pistas sugeridas

```python
# Para crear la playlist se requieren únicamente los id's

# Creamos 2 listas:
ids_t20 = []
ids_playlist = []

# Iterar sobre las top20
for i in range(top20_df.shape[0]):
    # EL nombre las canciones quedaron en el indice 0
    print(top20_df.index[i])
    # Agregamos el inidice a la lista
    ids_t20.append(top20_df['id'][i])
    
    # Obtener listado de candidatos para esta pista
    cands = obtener_candidatos(i, cos_sim, 5, umbral=0.8)
    
    # Si hay pistas relaciones obtener los ids correspondientes
    if len(cands) == 0:
        # e imprimir en pantalla que no se encontraron
        print('    *** No se encontraron pistas relacionadas***')
    else:
        # Obtener los ids correspondinetes e imprimir en pantalla
        for j in cands:
            # Asignar el id de la cancion segun el numero en la lista. Example: pista 154 == id
            id_cand = candidatos_df['id'][j]
            # Agregar id a la lista
            ids_playlist.append(id_cand)
            
            # Imprimir el candidato segun el iterador en el indice
            print(f'   {candidatos_df.index[j]}')
```

```python
# Eliminar candidatos que ya están en el top-20
# Se agregan a la lista todas las x que que no este en ids_20
ids_playlist_dep = [x for x in ids_playlist if x not in ids_t20]

# Eliminar posibles repeticiones
ids_playlist_dep = list(set(ids_playlist_dep))
```
